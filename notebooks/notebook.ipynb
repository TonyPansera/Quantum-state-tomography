```python
import pandas as pd

df = pd.read_csv("../data/qst_regression_dataset.csv")

df.head()
```

```python
import matplotlib.pyplot as plt

for col in ["X_mean", "Y_mean", "Z_mean", "theta", "phi", "theta_ideal", "phi_ideal"]:
    plt.figure()
    plt.hist(df[col].dropna(), bins = 50)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title(f"Histogram of {col}")
    plt.show()
```
